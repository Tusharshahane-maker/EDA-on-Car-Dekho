```python
# Get the value counts directly from the original DataFrame
cars_count = df["name"].value_counts().sort_values(ascending=False).head(20).reset_index()
# Rename columns to make them more intuitive
cars_count.columns = ['carnames', 'count']  
# Now use the correct column names in the px.bar() function
fig = px.bar(cars_count, x='carnames', y='count', title='Top 10 Most Common Cars')
fig.show()
```

```python
fuel=df.groupby(by="fuel")
```

```python
# First get the mean selling price by fuel type
fSell = fuel.selling_price.mean().reset_index()

# Rename columns to make them more intuitive
fSell.columns = ['fuel_type', 'average_price']  

# Sort the values in descending order
fSell = fSell.sort_values('average_price', ascending=False)

# Now use the correct column names in the px.bar() function
fig = px.bar(fSell, x='fuel_type', y='average_price', title='Fuel Wise Sell')
fig.show()
```

Disel Car have highest average selling price

```python
ndf=df.loc[~df['brand'].isin(top_brands)]
```

```python
print("Share of Normal Segment Cars in %: ") 
print(len(ndf)/len(df)*100)
print("-------------------------------------")
print("Share of premium Segment Cars in % : ") 
print(len(tdf)/len(df)*100)
```

```python
sns.relplot(x=ndf.selling_price,y=ndf.km_driven,hue=ndf.brand)
plt.show()
```

```python
brand_counts = ndf.brand.value_counts().reset_index()
brand_counts.columns = ['brand', 'count']

# Sort the DataFrame
brand_counts = brand_counts.sort_values(by='count', ascending=False).head()

# Create the bar plot using plotly express
fig = px.bar(brand_counts, 
             x='brand', 
             y='count',
             title='Brand Counts')

# Update layout for better readability
fig.update_layout(
    xaxis_title='Brand',
    yaxis_title='Count',
    xaxis={'categoryorder':'total descending'}
)

# Show the plot
fig.show()
```

Maruti & Hyundai are most popular brands in Normal Car Segment
